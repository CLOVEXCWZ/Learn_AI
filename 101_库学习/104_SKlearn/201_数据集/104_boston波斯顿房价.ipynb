```python
from sklearn.datasets import load_boston
import pandas as pd
import numpy as np
```

# data

    这个数据集包含了506处波士顿不同地理位置的房产的房价数据（因变量），和与之对应的包含房屋以及房屋周围的详细信息（自变量），其中包含城镇犯罪率、一氧化氮浓度、住宅平均房间数、到中心区域的加权距离以及自住房平均房价等13个维度的数据，因此，波士顿房价数据集能够应用到回归问题上，这里使用
    
- data
- target
- feature_names

```python
data = load_boston()
df = pd.DataFrame(data = data.data)
df.columns = data.feature_names
df['price'] = data.target
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
plt.figure(figsize = (12, 10))

columns = df.columns


for i in range(4):
    for j in range(4):
        num = i * 4 + j + 1 
        if num < len(columns) : 
            ax  = plt.subplot(4, 4, num)
            cl = columns[num - 1]
            sns.stripplot(x='price' , y=df[cl], data=df, jitter=False, ax = ax)


# ax2 = plt.subplot(2,2,2)
# ax3 = plt.subplot(2,2,3)
# ax4 = plt.subplot(2,2,4)
```
