## 1 .loc()
Pandas提供了各种方法来完成基于标签的索引。 切片时，也包括起始边界。整数是有效的标签，但它们是指标签而不是位置。
.loc()具有多种访问方式，如 

- 单个标量标签
- 标签列表
- 切片对象
- 一个布尔数组

```python
#------------- loc需要两个单/列表/范围运算符，用","分隔。第一个表示行，第二个表示列。 

# 示例-1
import pandas as pd
import numpy as np
df = pd.DataFrame(np.random.randn(8, 4),
index = ['a','b','c','d','e','f','g','h'], columns = ['A', 'B', 'C', 'D']) 
"""
print (df.loc[:,'A'])   # 只去出  'A' 这一列  以及 所有的行

a    0.315376
b   -1.180074
c   -1.455336
d   -0.081164
e    2.154814
f   -0.079533
g   -1.021639
h   -2.446844
Name: A, dtype: float64
"""

# 实例-2
"""
print (df.loc[:,['A','C']])  # 取出所有的行 以及 'A' 'C' 列

          A         C
a  0.353795  0.252048
b  0.014605  0.703627
c  0.171285 -0.587850
d  0.819027  0.627506
e -0.057000  1.670983
f  0.548817 -0.400891
g  0.022223 -0.205787
h -0.486299  0.357938
"""

# 示例-3
"""
print (df.loc[['a','b','f','h'],['A','C']])   # 取出a,b,f,h行 以及 A,C列

          A         C
a -0.721270  0.489307
b  0.158494  0.002560
f -0.890797  2.249173
h -0.046287 -0.068828
"""

# 示例-4
"""
print (df.loc['a':'g'])  # 取出a~g的行 以及所有的列

          A         B         C         D
a -1.178457  0.056994  0.825480 -0.203953
b  0.140499 -0.749409 -1.162640 -0.734937
c -0.534302  0.298742  0.573491 -0.160175
d -1.077197  1.275537 -1.655574 -0.325382
e  0.490867  0.222272  0.179172 -0.057415
f  1.247118 -0.039506 -0.655504 -1.274760
g -0.502419  1.954480 -0.197085 -0.152705
"""

# 示例 - 5
"""
print (df.loc['a']>0)  # 获取行a是否大于0 (布尔值列表)

A     True
B    False
C     True
D    False
Name: a, dtype: bool
"""
print()
```

## 2 .iloc()
Pandas提供了各种方法，以获得纯整数索引。像python和numpy一样，第一个位置是基于0的索引。
各种访问方式如下

- 整数
- 整数列表
- 系列值 

```python
import pandas as pd
import numpy as np
df = pd.DataFrame(np.random.randn(8, 4), columns = ['A', 'B', 'C', 'D']) 
"""
print (df.iloc[:4])  # [0, 4）的行 以及所有的列

         A         B         C         D
0  1.242026  1.121167  0.635715 -0.298944
1 -0.250900  1.080250  0.668667 -0.706112
2 -0.472993 -0.189758 -0.787341  0.942780
3 -0.437530 -0.293697 -1.090370  1.284850
"""

"""
print (df.iloc[1:5, 2:4]) # [1, 5)的行 以及 [2, 4)的列

          C         D
1  1.073342 -1.856620
2 -0.762715 -0.872655
3 -0.767101  0.470352
4 -0.255323 -1.346814
"""

print (df.iloc[[1, 3, 5], [1, 3]]) # 第1，3，5行  1，3列
print (df.iloc[1:3, :]) # [1, 3)行 以及所有列
print (df.iloc[:,1:3])  # 所有行，[1, 3)列

"""
          B         D
1  0.132129  0.420515
3 -0.064392  0.042074
5 -1.258598  0.138642
          A         B         C         D
1 -1.268063  0.132129  1.396815  0.420515
2 -0.824232  1.757488  0.467758 -0.458639
          B         C
0  0.807193 -1.618589
1  0.132129  1.396815
2  1.757488  0.467758
3 -0.064392 -0.130710
4 -0.989877  0.752334
5 -1.258598 -1.275066
6  0.716868  1.256211
7  1.380730 -1.425501
"""
```

## 4 直接访问

```python
import pandas as pd
import numpy as np
df = pd.DataFrame(np.random.randn(8, 4), columns = ['A', 'B', 'C', 'D'])
print(df)
print (df['A'])  # 直接返回 A 的数据
print (df[['A','B']]) # 返回A、B 列的数据
print (df[0:2]) #返回0~2行
```

## 属性访问
可以使用属性运算符.来选择列

```python
import pandas as pd
import numpy as np
df = pd.DataFrame(np.random.randn(8, 4), columns = ['A', 'B', 'C', 'D'])
print (df.A) 
```
