```python
from sklearn import datasets
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = datasets.make_blobs(n_samples=1000, 
                           n_features=2, 
                           centers=4, 
                           cluster_std=1.0, 
                           center_box=(-10.0, 10.0), 
                           shuffle=True, 
                           random_state=None)
```

```python
feature, labels = data 
df = pd.DataFrame(data = feature)
df.columns = ['a', 'b']
df['label'] = labels 
df.head()
```

```python
import seaborn as sns
```

```python
sns.pairplot(df[['a', 'b',  'label']], hue="label", diag_kind = 'kde')
```

```python
sns.FacetGrid(df, hue = 'label', height = 5.0).map(plt.scatter, 'a', 'b').add_legend()
```

```python
plt.scatter(df.a,df.b,c=df.label,s=8)
```
