```python
from keras_bert import get_base_dict, get_model, gen_batch_inputs
import os
import codecs
```

```python
# 定义地址
bert_model_path = "/Users/zhouwencheng/Desktop/Grass/data/model/ImportModel/BERT/chinese_L-12_H-768_A-12"
config_path = os.path.join(bert_model_path, 'bert_config.json')
checkpoint_path = os.path.join(bert_model_path, 'bert_model.ckpt')
vocab_path = os.path.join(bert_model_path, 'vocab.txt')
```

```python

os.environ['TF_KERAS'] = '1'
token_dict = {}
with codecs.open(vocab_path, 'r', 'utf8') as reader:
    for line in reader:
        token = line.strip()
        token_dict[token] = len(token_dict)
```

```python
model = get_model(
    token_num=len(token_dict),
    head_num=128,
    transformer_num=12,
    embed_dim=768,
    feed_forward_dim=100,
    seq_len=512,
    pos_num=512,
    dropout_rate=0.05, 
) 
model.summary()
```

```python
# 把模型保存为图片
from keras.utils import plot_model
plot_model(model,to_file='BERT.png', show_layer_names=True, show_shapes=True)
```
