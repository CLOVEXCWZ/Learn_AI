```python
def cut_content(contents):
    cut_content = [] 
    for line in contents:
        cut_line = jieba.lcut(line) 
        cut_content.append(cut_line) 
    return cut_content 
cut_contens = cut_content(df.content) 
```

# 去掉停词

```python
# 获取停词

stop_path = "/Users/zhouwencheng/Desktop/Grass/05 AI/06 data/10 data/01 文本/stopwords.txt"
stopwords = pd.read_csv(stop_path, index_col=False, sep="\t",quoting=3,names=['stopword'], encoding='utf-8')
stopwords.head() 
```

```python
def cut_of_stop_word(sentence, stop_words):
    content_s = []
    all_words = []
    for line in df.content:
        sentence = []
        for word in line:
            if word not in stop_words:
                sentence.append(word)
                all_words.append(word)
        content_s.append(sentence)
    return content_s, all_words
        
stop_words = stopwords.stopword.values.tolist()
content_s, all_words = cut_of_stop_word(df.content, stop_words)
df.content = content_s

df.head()
```
