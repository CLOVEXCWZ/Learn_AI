    数据不均衡才去下采样的方法

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/Users/zhouwencheng/Desktop/Grass/05 AI/05 jupyter/10 data/creditcard.csv")
df.head()
```

```python
# 数据上看，欺诈的数据是很少的 (492)  正常数据是很多的(284807) 
print('shape:', df.shape)
df.Class.value_counts()
```

```python
# 先对数据进行归一下处理

from sklearn.preprocessing import StandardScaler  

df.Amount = StandardScaler().fit_transform(df.Amount.values.reshape(-1, 1))
df.head()
```

# 下采样

```python
fraud_indices = df.Class[df.Class == 1].index
fraud_count = len(fraud_indices)

normal_indices = df.Class[df.Class == 0].index

random_normal_indices = np.random.choice(normal_indices,  fraud_count, replace=False)# 随机从正常数据中选出欺骗数据的数量的数据的索引值

under_sample_indices = np.concatenate([fraud_indices, random_normal_indices])
under_sample_data = df.loc[under_sample_indices]

# 这便是下采样的数据了
under_sample_data
```
