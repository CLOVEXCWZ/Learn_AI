```python
import pandas as pd
import numpy as np
import matplotlib.pylab
%matplotlib inline 
```

# 载入数据(创建数据)

```python
df = pd.Series(np.random.randn(600), index = pd.date_range('7/1/2016', freq = 'D', periods = 600))
df.head()
```

# 滑动窗口

```python
# r = df.rolling(window = 10, min_periods=1)
# r.mean()

r = df.rolling(window = 10) # 滑动10个 
r.mean().head(10)  # 算平均值
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize = (15, 5))

df.plot(style='r--')
df.rolling(window=10).mean().plot(style='b')
```
