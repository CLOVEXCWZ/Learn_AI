```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
```

```python
df = pd.read_csv("/Users/zhouwencheng/Desktop/Grass/05 AI/05 jupyter/10 data/creditcard.csv")
df.head()
```

```python
# 先对数据进行归一下处理

from sklearn.preprocessing import StandardScaler  

df.Amount = StandardScaler().fit_transform(df.Amount.values.reshape(-1, 1))
df.head()
```

```python
features = df.loc[:, df.columns != 'Class']
labels    = df.loc[:, df.columns == 'Class']
features.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size = 0.3, random_state = 0)
print('Class0 = ', len(y_train.Class[y_train.Class == 0]))
print('Class1 = ', len(y_train.Class[y_train.Class == 1]))
X_train.head()
```

# 过采样操作

```python
oversampler = SMOTE(random_state = 0)
os_features, os_labels = oversampler.fit_sample(X_train, y_train.values.ravel()) 

print('Class0 = ', len(os_labels[os_labels == 0]))
print('Class1 = ', len(os_labels[os_labels == 1])) 

# 转化成DataFrame结构
# os_features = pd.DataFrame(os_features)
# os_labels = pd.DataFrame(os_labels)
```
